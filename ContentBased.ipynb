```python
import sklearn
from sklearn.preprocessing import MinMaxScaler
```

```python
num_cols=ini_df.select_dtypes(include='number').columns
num_cols=num_cols.drop(labels='year')
num_cols
```

```python
scaler=MinMaxScaler()
scaler.fit(ini_df[num_cols])
scaler.get_params()
```

```python
ini_df[num_cols]=scaler.transform(ini_df[num_cols])
ini_df[num_cols].describe()
```

```python
tag_df = pd.DataFrame(ini_df[['track_id','tags']])
tag_df["tags_id"] = tag_df["tags"].astype("category").cat.codes
tag_df['tags_id'].nunique()
```

```python
import torch
import torch.nn as nn

num_categories = tag_df["tags_id"].nunique() 
embedding_dim = 5  

embedding_layer = nn.Embedding(num_categories, embedding_dim)

category_indices = torch.tensor(tag_df["tags_id"].values, dtype=torch.long)

embedded_features = embedding_layer(category_indices)

embedded_features.shape
```

```python
embedded_features
```

```python
embedding_df = pd.DataFrame(embedded_features.detach().numpy(), columns=[f"tags_emb_{i}" for i in range(5)])

df = pd.concat([ini_df.drop(columns=["tags"]), embedding_df], axis=1)
```

```python
df
```

```python
artist_counts = df['artist'].value_counts()
df['artist_encoded'] = df['artist'].map(artist_counts)
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df
```

```python
from sklearn.neighbors import NearestNeighbors

drop_cols=['track_id','name','artist']
features=df.drop(columns=drop_cols).columns.tolist()
feature_matrix = df[features].values
knn = NearestNeighbors(n_neighbors=5, metric="cosine")
knn.fit(feature_matrix)
```

```python
def knn_recommend_by_name(song_name, df, feature_matrix, knn):

    try:
        idx = df[df["name"] == song_name].index[0]  
        distances, indices = knn.kneighbors([feature_matrix[idx]])
        recommended_track_ids = df.iloc[indices[0][1:]]["track_id"].tolist()
        recommended_song_names = df[df["track_id"].isin(recommended_track_ids)]["name"].tolist()
        return recommended_song_names

    except IndexError:
        print(f"Song '{song_name}' not found in the dataset.")
        return None
    except KeyError:
        print("Required columns 'name' or 'track_id' not found in DataFrame")
        return None
    except Exception as e:
        print(f"An error occurred: {e}")
        return None
```

```python
knn_recommend_by_name("God Is A Girl",df,feature_matrix,knn)
```

```python
import joblib
joblib.dump(knn, "content_model.pkl")
print("Model saved successfully.")
```
